#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
plt.plot(x, y, label='y')
plt.plot(x, z, label='z')
plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT') 
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.fig

ax2.plot(x,z)
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

plt.show()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(10, 5))
ax1.plot(x,y)
ax1.set_title('LINEAR PLOT') 
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,z)
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
xaxis = np.linspace(0, 4, 1000)
y1 = lambda x: x**2
y2 = lambda x: np.exp(x)

plt.plot(xaxis, y1(xaxis), color= 'k', lw= 2, label= 'x^2')
plt.plot(xaxis, y2(xaxis), color= 'r', lw= 2, label= 'e^x')
plt.legend()
```
